## Condicionales

### Si-Sino-Finsi
Recuerda que los condicionales nos permiten decidir qué instrucciones ejecutar y cuales no. El condicional Si-Sino-Finsi tiene la siguiente estructura:

    Si (condición) Entonces
        instrucción(es)
    Sino
        instrucción(es)
    Fin Si

En Python, esto se escribe un poco diferente y la estructura general depende de las tabulaciónes. Por ejemplo:

```python
x = int(input("Por favor ingresa un número: "))
if x<0:
    print('X es Negativo')
else:
    print('X es Positivo')

print('Esto ya está por fuera del condicional') #Esta fuera porque ya no hay tabulación
```

¿Qué pasa si X es 0? ¿Imprimirá positivo o negativo? ¿Cómo podríamos arreglar este programa para que si X es cero, imprima "X es cero"?



```python
x = int(input("Por favor ingresa un número: "))
if x<0:
    print('X es Negativo')
elif x == 0:
    print('X es cero')
else:
    print('X es Positivo')
```

# Ejercicios con condicionales #1.   

Diseñe un algoritmo que lea tres longitudes y determine si forman o no un triángulo. Si es un triángulo determine de qué tipo de triángulo se trata entre: equilátero (si tiene tres lados iguales), isósceles (si tiene dos lados iguales) o escaleno (si tiene tres lados desiguales). Considere que para formar un triángulo se requiere que: "el lado mayor sea menor que la suma de los otros dos lados". Algoritmo

```python
lado_a = int(input("Lado A: "))
lado_b = int(input("Lado B: "))
lado_c = int(input("Lado C: "))

if (lado_a < (lado_b + lado_c)) and (lado_b < (lado_a + lado_c)) and (lado_c < (lado_a + lado_b)):
  print("Es un triángulo.")
  if lado_a == lado_b == lado_c:
    print("Es un triángulo equilátero.")
  elif lado_a == lado_b or lado_b == lado_c or lado_a == lado_c:
    print("Es un triángulo isósceles.")
  else:
    print("Es un triángulo escaleno.")
else:
  print("No es un triángulo.")
```

# Ejercicio con condicionales #3

Se desea escribir un algoritmo que pida la altura de una persona, si la altura es menor o igual a 1.50 cm envíe el mensaje: Persona de altura baja; si la altura está entre 1.51 y 1.70 escriba el mensaje: Persona de altura media y si la altura es mayor al 1.71 escriba el mensaje: Persona alta. Exprese el algoritmo en código python. 

```python
altura = float(input("Ingrese la altura en metros: "))

if altura <= 1.5:
  print("Persona de altura baja.")
elif altura > 1.5 and altura <= 1.7:
  print("Persona de altura media.")
else:
  print("Persona alta.")
```

# Ejercicio con condicionales #4

Un empleado necesita calcular su salario semanal, el cual se obtiene de la sig. manera:
Si trabaja 40 horas o menos se le paga 16 por hora
Si trabaja mas de 40 horas se le paga 16 por cada una de las primeras 40 horas y 20 por cada hora extra.

```python
horas = int(input("Ingrese las horas trabajadas en la semana: "))
if horas <= 40:
  pagado = horas * 16
else:
  extras = horas - 40
  pagado = (40 * 16) + (extras * 20)
print("El salario recibido en la semana es: $", pagado)
```

# Ejercicio con condicionales #5

Hacer un algoritmo que calcule el total a pagar por la compra de camisas. Si se compran tres camisas o mas se aplica un descuento del 25% sobre el total de la compra y si son menos de tres camisas un descuento del 10%

```python
precio_camisa = int(input("Precio por camisa: $ "))
cantidad = int(input("Cantidad a comprar: "))

if cantidad < 3:
  descuento = 0.1
else:
  descuento = 0.25

total_a_pagar = cantidad * (precio_camisa * (1 - descuento))
print("El total a pagar es: $", total_a_pagar)
```
